### Standardization 

The formula for Z-score standardization is:

z = (x - μ) / σ

Where:

- z is the standardized value.
- x is the original value.
- μ is the mean of the feature.
- σ is the standard deviation of the feature.

This is done using the **StandardScaler** from the preprocessing sklearn library of python

PCA is a 
statistical analysis that allows the reduction of a large multivariate set of vulnerability indicators 
into principal components while retaining key statistical information and spatial patterns (Jollife, 
1986). These components have loading values that are associated with each of the 
vulnerability indicators in the data set. The loadings portray the type (positive or negative) and 
strength of the relationship between a principal component and an indicator and thus provide 
information about the patterns of vulnerability within the data that each component is likely to 
represent. PCA is often referred to as a global data reduction technique whereby it produces 
one set of components for the whole data set, representing the whole of Amsterdam in this 
research. To get an understanding of the spatial distribution of the vulnerability that each 
principal component and locales in which the vulnerability is likely to enhance can be mapped 
by the component loadings. However, these loadings are spatially stationary and thus cannot 
account for spatial effects (Openshaw, 1984; Lloyd, 2010; Harris et al., 2011; Demšar et al., 
2013). The PCA is carried out in IBM’s SPSS program.

### Bartlett's Test of Sphericity

Bartlett's test of sphericity is a statistical test used to determine whether the correlation matrix of your variables is an identity matrix, indicating that your variables are not suitable for factor analysis. If Bartlett's test is statistically significant (i.e., the p-value is less than a chosen significance level, often 0.05), then you reject the null hypothesis, indicating that the correlation matrix is not an identity matrix and that factor analysis may be appropriate. 

```python
from factor_analyzer.factor_analyzer import calculate_bartlett_sphericity
from scipy.stats import chi2_contingency

chi_square_value, p_value = calculate_bartlett_sphericity(df1)
degrees_of_freedom = df1.shape[1] * (df1.shape[1] - 1) // 2

print("Degrees of freedom:", degrees_of_freedom)
print("Chi-square value:", chi_square_value)
print("Sig. P-value:", p_value)
```

In our case the **Sig. (p-value) is 2.54688009777057e-190**, therefore we reject the null hypothesis and conclude that our correlation matrix not an identity matrix and factor analysis is appropriate for the data 

### Principal Component Analysis

- It is a dimensionality reduction technique. 
- It uses a method called SSD (sum of squared distances) to calculate the Principal Components. (Each PC has a variance (% of the data covered).
- From each PC, eigen vectors can be calculated (per unit). 
- The loading scores (tell you the impact that each variable has on the PC and the cardinality (+/-). 
- The Eigen values are the average SSD/n-1 of the PCs. 

```python
from factor_analyzer import FactorAnalyzer
# Instantiate the FactorAnalyzer object
fa = FactorAnalyzer(n_factors=34, rotation='varimax', method='mle')

# Fit the FactorAnalyzer object to your data
fa.fit(df1)
```

### Communalities

**What are communalities?** <br> Communalities refer to:
- the proportion of variance (common variance) of each variable being represented by the PCs and ranges between 0 - 1. 
- Value closer to 1 indicates that the extracted PCs explain more variance of this variable.(For ex. (%65 year old and above has a communality of 0.995 which means the PCs 

```python
factor_communalities = fa.get_communalities()

comm_data = {
    'Variable': df1.columns,
    'Communality': factor_communalities
}
comm_df = pd.DataFrame(comm_data)

comm_df
```

### Eigenvalues and explained variance

**What are Eigen Values?** <br> Eigenvalues represent the amount of variance that is explained by a given Principle Component

```python
eigenvalues = fa.get_eigenvalues()[0]

total_variance = np.sum(eigenvalues)

percentage_variance = eigenvalues / total_variance * 100

cumulative_percentage = np.cumsum(percentage_variance)

print(eigenvalues)
```

```python
eigenvalues_table = {
    'Component': np.arange(1, len(eigenvalues) + 1),
    'Eigenvalue': eigenvalues,
    '% Variance': percentage_variance,
    'Cumulative %': cumulative_percentage
}

eigen_df = pd.DataFrame(eigenvalues_table)
```
